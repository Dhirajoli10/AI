```
# TASK 3 Q1
x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

#ADD
add_arr = x + y
print(f"Add array : \n {add_arr}")

#SUBTRACT
sub_arr = x - y
print(f"Subtract array : \n {sub_arr}")


#MULTIPLY
mulX_arr = x * 2
mulY_arr = y * 3
print(f"Multiply X array : \n {mulX_arr}")
print(f"Multiply Y array : \n {mulY_arr}")


#Square of each element of array
sqrX_arr = np.square(x)
sqrV_arr = np.square(v)
print(f"Sqaure of X: \n {sqrX_arr}")
print(f"Sqaure of V: \n {sqrV_arr}")


#dot product of v and w
dotPofvNw = np.dot(v,w)
print(f"Dot product of v and w: \n {dotPofvNw}")

#dot product of x and v
dotPofxNv = np.dot(x,v)
print(f"Dot product of v and w: \n {dotPofxNv}")


#dot product of x and y
dotPofxNy = np.dot(x,y)
print(f"Dot product of v and w: \n {dotPofxNy}")

#Concatenate x and y in row and Concatenate v and w in column
conarr0 = np.concatenate((x,y), axis = 0)
conarr1 = np.vstack((v,w)).T

print(f"Concatenate x and y \n {conarr0}")
print(f"Concatenate v and w \n {conarr1}")
```

```
# Task 4 Q1
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# Prove A.Aâˆ’1 = I

#Finding A.A-1
ainve = np.linalg.inv(A)
result1 = np.dot(A, ainve)
print(result1)
print(f"\n")


#Q1
#2
# To prove AB /= BA
# Finding AB

ab = np.dot(A,B)
print(ab)

ba = np.dot(B,A)
print(ba)

print(f"AB is not equal to BA")
print(f"\n")
#Q1
#3
# To Prove (AB)T = BTAT

#finding (AB)T

abTrans = ab.T
print(f"(AB)T  = \n {abTrans}")

#finding BTAT

btrans = B.T

atrans = A.T

btat = np.dot(btrans, atrans)
print(f"BTAT  = \n {btat} \n ")

print(f"(AB)T  = BTAT")
```
